# February 2023 Challenge: Mentorship Program Screening Tasks

Since this month coincides with the application window for the QOSF mentorship program, we have decided to make the corresponding screening tasks the challenge for this month. So, even if you aren't applying to the mentorship program you can still have a look at the screening tasks.

There are four separate tasks, so you're welcome to try your hand at as many of them as you like!

**_NOTE: The application deadline is February 28th, so don't submit any PRs until after that date since you would be making potential solutions public.**

## Task 3 QSVM

Generate a Quantum Support Vector Machine (QSVM) using the [iris dataset](https://archive.ics.uci.edu/ml/datasets/iris) and try to propose a kernel from a parametric quantum circuit to classify the three classes(setosa, versicolor, virginica) using the one-vs-all format, the kernel only works as binary classification. Identify the proposal with the lowest number of qubits and depth to obtain higher accuracy. You can use the UU† format or using the [Swap-Test](https://en.wikipedia.org/wiki/Swap_test).
